```python
# These work the best
from numba.extending import intrinsic


@intrinsic
def bitcast_float64_to_int64(typingctx, x):
    sig = nb.int64(x)

    def codegen(context, builder, signature, args):
        [x_val] = args
        return builder.bitcast(x_val, context.get_value_type(nb.int64))

    return sig, codegen


@intrinsic
def bitcast_int64_to_float64(typingctx, x):
    sig = nb.float64(x)

    def codegen(context, builder, signature, args):
        [x_val] = args
        return builder.bitcast(x_val, context.get_value_type(nb.float64))

    return sig, codegen
```

```python
def diamond_transform(img):
    """
    Applies the diamond edge transformation using plain NumPy vectorization.

    Mapping Logic derived from user example:
    - Horizontal Diffs at (y, x): u = y + x, v = x - y + (H - 1)
    - Vertical Diffs at (y, x):   u = y + x, v = x - y + (H - 1) - 1
    """
    # Ensure input is an array
    img = np.asarray(img)
    H, W = img.shape

    # 1. Calculate Differences (Vectorized)
    # diff_h: difference between column x and x+1. Shape (H, W-1)
    diff_h = np.abs(img[:, :-1] - img[:, 1:])

    # diff_v: difference between row y and y+1. Shape (H-1, W)
    diff_v = np.abs(img[:-1, :] - img[1:, :])

    # 2. Generate Coordinate Grids
    # Indices for Horizontal Diffs
    # rows_h ranges 0..H-1, cols_h ranges 0..W-2
    rows_h, cols_h = np.indices((H, W - 1))

    # Indices for Vertical Diffs
    # rows_v ranges 0..H-2, cols_v ranges 0..W-1
    rows_v, cols_v = np.indices((H - 1, W))

    # 3. Calculate Target Coordinates
    # Offset ensures the minimum v-index starts at 0.
    offset = H - 1

    # Map Horizontal Diffs
    u_h = rows_h + cols_h
    v_h = cols_h - rows_h + offset

    # Map Vertical Diffs
    # Vertical diffs sit "to the left" of the corresponding horizontal diff in the rotated grid
    u_v = rows_v + cols_v
    v_v = cols_v - rows_v + offset - 1

    # 4. Create Output Canvas
    # The output shape is determined by the max range of u and v.
    # Max u = (H-1) + (W-1) - 1 approx. Exact size is H + W - 2.
    out_size = H + W - 2

    # Handle edge case for very small images (1x1)
    if out_size <= 0:
        return np.zeros((0, 0), dtype=img.dtype)

    canvas = np.zeros((out_size, out_size), dtype=img.dtype)

    # 5. Assign Values (Vectorized Advanced Indexing)
    # We check bounds just in case, though the math guarantees fit for H,W > 1
    valid_h = (u_h < out_size) & (v_h < out_size) & (v_h >= 0)
    canvas[u_h[valid_h], v_h[valid_h]] = diff_h[valid_h]

    valid_v = (u_v < out_size) & (v_v < out_size) & (v_v >= 0)
    canvas[u_v[valid_v], v_v[valid_v]] = diff_v[valid_v]

    return canvas


a = np.asarray(
    [
        [1, 2, 4],
        [8, 16, 32],
        [64, 128, 256],
    ]
)
diamond_transform(a)
```

```python
@njit
def weighted_random_index(weights):
    """
    Select index i with probability proportional to weights[i].
    Mutates `weights` in-place into prefix sums.
    Uses searchsorted with side='right' so we pick the smallest i with prefix[i] > r.
    """
    n = weights.shape[0]
    if n == 0:
        raise ValueError("weights must be non-empty")

    # build prefix sums in-place
    for i in range(1, n):
        weights[i] += weights[i - 1]

    total = weights[n - 1]
    if total <= 0.0:
        raise ValueError("sum of weights must be positive")

    # sample r in [0, total)
    r = np.random.random() * total

    # return smallest i with prefix[i] > r  matches half-open intervals [low, high)
    return np.searchsorted(weights, r, side='right')
```

```python
import numba
import numpy as np
import time


# -------------------------
# Inline-aware decorator (fixed)
# -------------------------
def njit_inline(use_inline=True):
    # decorator factory
    def decorator(func):
        if use_inline:
            return numba.njit(inline='always')(func)
        else:
            return numba.njit(func)

    return decorator


# -------------------------
# Example helpers
# -------------------------
def define_helpers(use_inline=True):
    @njit_inline(use_inline)
    def helper1(x):
        return x * x + 1

    @njit_inline(use_inline)
    def helper2(x):
        return helper1(x) + np.sin(x)

    @njit_inline(use_inline)
    def helper3(x):
        return helper2(x) + np.cos(x)

    return [helper1, helper2, helper3]


# -------------------------
# Benchmarking wrapper
# -------------------------
def benchmark_helpers(helpers, n_iter=10_000_000):
    arr = np.linspace(0, 10, n_iter, dtype=np.float64)
    results = []

    for h in helpers:
        # Top-level loop calling the helper
        @numba.njit
        def hot_loop(arr):
            s = 0.0
            for x in arr:
                s += h(x)
            return s

        # Warmup (trigger compilation)
        hot_loop(arr[:10])

        # Timing
        start = time.perf_counter()
        hot_loop(arr)
        elapsed = time.perf_counter() - start

        results.append((h.__name__, elapsed))

    return results


# -------------------------
# Run benchmark
# -------------------------
def run_benchmark():
    for use_inline in [True, False]:
        helpers = define_helpers(use_inline)
        print(f"\n=== USE_INLINE={use_inline} ===")
        results = benchmark_helpers(helpers)
        for name, t in results:
            print(f"{name}: {t:.4f} seconds")


# -------------------------
# Execute
# -------------------------
if __name__ == "__main__":
    run_benchmark()
```

```python
@njit
def weighted_random_index(rng, weights):
    """
    Select index i with probability proportional to weights[i].
    Mutates `weights` in-place into prefix sums.
    Uses searchsorted with side='right' so we pick the smallest i with prefix[i] > r.
    """
    n = len(weights)
    if n == 0:
        raise ValueError("weights must be non-empty")

    # build prefix sums in-place
    for i in range(1, n):
        weights[i] += weights[i - 1]

    total = weights[n - 1]
    if total <= 0.0:
        raise ValueError("sum of weights must be positive")

    # sample r in [0, total)
    r = rng.random() * total

    # return smallest i with prefix[i] > r  matches half-open intervals [low, high)
    return np.searchsorted(weights, r, side='right')


@njit
def custom_weighted_random_index(rng, weights, power):
    if power == 0.0:
        return rng.integers(0, len(weights))
    if np.isfinite(power):
        weights **= power
        return weighted_random_index(rng, weights)
    else:
        if power == np.inf:
            target_weight = weights.max()
        elif power == -np.inf:
            target_weight = weights.min()
        else:
            raise ValueError("power must be finite or infinite")
        for i in range(weights.shape[0]):
            weights[i] = weights[i] == target_weight
        return weighted_random_index(rng, weights)
```

```python
int_dtype = np.int64


@njit
def top_level(length):
    width = 8
    size = length * width
    buffer = np.empty(width, dtype=int_dtype)
    a = np.arange(size - 1, -1, -1).reshape((length, width))
    out = np.empty(length, dtype=int_dtype)
    for i in range(length):

        jj = np.argsort(a[i])
        a[i] = a[i, jj]

        buffer[0] = a[i, 0]
        for j in range(1, width):
            buffer[j] = buffer[j - 1] + a[i, j]
        r = np.random.random() * buffer[-1]
        out[i] = np.searchsorted(buffer, r, side='right')


@njit
def with_rng(rng, length):
    width = 8
    size = length * width
    buffer = np.empty(width, dtype=int_dtype)
    a = np.arange(size - 1, -1, -1).reshape((length, width))
    out = np.empty(length, dtype=int_dtype)
    for i in range(length):

        jj = np.argsort(a[i])
        a[i] = a[i, jj]

        buffer[0] = a[i, 0]
        for j in range(1, width):
            buffer[j] = buffer[j - 1] + a[i, j]
        r = rng.random() * buffer[-1]
        out[i] = np.searchsorted(buffer, r, side='right')


rng = np.random.default_rng(0)
top_level(2)
with_rng(rng, 2)

length = 4096**2

t0 = time.time()
top_level(length)
print(time.time() - t0)

t0 = time.time()
with_rng(rng, length)
print(time.time() - t0)
```

```python
my_spec = (
    ('data', nb.float64[:]),
    ('rtol', nb.optional(nb.float64)),
    ('atol', nb.optional(nb.float64)),
)


@jitclass(my_spec)
class MyClass:
    def __init__(self, data, rtol=None, atol=None):
        self.data = np.asarray(data, dtype=np.float64)
        self.rtol = rtol
        self.atol = atol

    def sum(self):
        print("MyClass original")
        return self.data.sum()

    def query(self, x, rtol=None, atol=None):
        rtol = rtol if rtol is not None else self.rtol
        atol = atol if atol is not None else self.atol
        if rtol is not None:
            if atol is not None:
                return np.isclose(self.data, x, rtol=rtol, atol=atol)
        return False


my = MyClass(data=[4.0, 5.0], rtol=1e-5, atol=1e-8)
my.query(10.0)
```

```python
from pytil.new_utility import isclose

my_spec = (
    ('data', nb.float64),
    ('rtol', nb.float64),
    ('atol', nb.float64),
)

NOT_GIVEN = -1.0


@jitclass(my_spec)
class MyClass:
    def __init__(self, data):
        self.data = data
        self.rtol = 1e-9
        self.atol = 0.0

    def set_tolerance(self, rtol, atol):
        self.rtol = rtol
        self.atol = atol

    def sum(self):
        print("MyClass original")
        return self.data.sum()

    def query(self, x, rtol=NOT_GIVEN, atol=NOT_GIVEN):
        rtol = rtol if rtol != NOT_GIVEN else self.rtol
        atol = atol if atol != NOT_GIVEN else self.atol
        return isclose(self.data, x, rel_tol=rtol, abs_tol=atol)


my = MyClass(data=5.0)
my.set_tolerance(0.0, 1.0)
my.query(6.0, 0.0, 0.5)
```

```python
from numba import njit
import numpy as np

n = 5


@njit
def myOptimiser(x, y, opt1=None, opt2=None):
    if opt1 is None:
        if opt2 is None:
            return np.ones((n,)) + y
        else:
            return x + opt2
    else:
        if opt2 is None:
            return x * np.ones((n,)) + opt1
        return x - y


@njit
def call_opt(a, b):
    return myOptimiser(a, b, np.ones((n,)))


print(call_opt(10, 20))
```
